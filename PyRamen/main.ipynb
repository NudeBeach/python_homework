```python
# Import necessary libraries
import csv
from pathlib import Path

# Set file paths for menu and sales csv's
menu_filepath = Path('Menu_Data.csv');
sales_filepath = Path('Ramen_Sales_Data.csv');

# Initialize list objects for menu and sales data
menu = [];
sales = [];

# Read in the menu data into the menu list

with open(menu_filepath,'r') as menu_file:
    menu_reader = csv.reader(menu_file)
    menu_header = next(menu_reader)
    print(menu_header)
    for line in menu_reader:
        menu.append(line)

# Read in the sales data into the sales list

with open(sales_filepath,'r') as sales_file:
    sales_reader = csv.reader(sales_file)
    sales_header = next(sales_reader)
    #print(sales_header)
    for row in sales_reader:
        sales.append(row)
        
# Initialize dict object to hold key-value pairs of items and metrics
report = {}
# Initialize a row counter variable
row_count_sales = 0;
row_count_menu = 0;
# Loop over every row in the sales list object and build report
for x in sales:
    report[x[4]] = {"01-count":0, "02-revenue":0, "03-cogs":0, "04-profit":0}

# Use a nested for loop to populate data in report dictionary for matching sales and menu items    
for y in menu:
    row_count_menu += 1 
    for x in sales:
        row_count_sales += 1
        if y[0] in x:
            report[x[4]]["01-count"] += int(x[3])   # Calculate the total count for the items.
            report[x[4]]["02-revenue"] += int(y[3]) * int(x[3]) # Calculate the total revenue for the items.
            report[x[4]]["03-cogs"] += int(y[4]) * int(x[3]) # Calculate the total cost of goods and services. 
            report[x[4]]["04-profit"] += (int(y[3]) * int(x[3])) - (int(y[4]) * int(x[3])) # Calculate the final profit. 

# If the item is not in the report, add it as a new entry with initialized metrics
        #else:
            #report[y[0]] = {"01-count":0, "02-revenue":0, "03-cogs":0, "04-profit":0}
# Check work
print(report)
print(row_count_menu)
print(row_count_sales)

print(f'Ichiban Ramen Annual Sales Report')
print(f'-------------------------------------------')
print(f'{report}')
print(f'The total menu items for the year: {row_count_menu}')
print(f'The total sales for the year: {row_count_sales}')
print(f'WE DID IT!!!!')
      
# Write the contents of the report dictionary to a text file.
output_path = Path("ichiban_ramen_report.txt")
with open(output_path, 'w') as file_object:
        file_object.write(f'''
Ichiban Ramen Annual Sales Report
-------------------------------------------
{report}
The total menu items for the year: {row_count_menu}
The total sales for the year: {row_count_sales}
WE DID IT!!!!''')
    

      





```
